#### Ask a question which feature is going to give the maximum info gain and minimum GINI.

```python
class Ask:

    def __init__(self, column, value):
        self.column = column
        self.value = value
        
    def match(self, sample):
        value_s = sample[self.column]
        if is_numeric(value_s):
            return value_s >= self.value
        else:
            return value_s == self.value
    
    #Representation - only needed if some of the values are numeric. Format gets the values in the columns
    def __repr__(self):
        return Ask("Is {self.column} greater than {self.value}?")
```
