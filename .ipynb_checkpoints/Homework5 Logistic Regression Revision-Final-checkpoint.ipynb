### Logistic Regression

```python
X = weather_data.drop("Precipitation", axis = 1)
X
```

```python
y = weather_data["Precipitation"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=110)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train) 
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
weather_model = LogisticRegressionCV()
```

```python
weather_model.fit(scaled_X_train,y_train)
```

```python
weather_model.Cs_
```

```python
#C value with best fit
weather_model.C_
```

```python
weather_model.coef_
```

```python
c_list = list(weather_model.coef_[0])
coef_list = [round(x, 2) for x in c_list]
coef_list
```

```python
coeff = list(weather_model.coef_[0])
labels = list(X_train.columns)
features = pd.DataFrame()
features['Features'] = labels
features['importance'] = coeff
features.sort_values(by=['importance'], ascending=False, inplace=True)
features['negative'] = features['importance'] < 0
features.set_index('Features', inplace=True)
features.importance.plot(kind='barh', figsize=(11, 6),color = features.negative.map({True: 'blue', False: 'red'}))
plt.xlabel('Importance')
plt.show()
```

The weather_model.predict displays the likelihood of rain for each combination of feature coefficients on the X_test.
The weather_model.predict.proba indicates the probability that there would be precipitation for each combination of feature coefficients, i.e if Temp_Avg rises by 1 unit, the probability of precipitation decreases, if Dew_Pt_Average increases by 1 unit, the probability of precipitation increases, and so on for each feature.¶

```python
y_pred = weather_model.predict(scaled_X_test)
```

```python
y_pred
```

```python
y_pred_pr = weather_model.predict_proba(scaled_X_test)
```

```python
y_pred_pr
```
